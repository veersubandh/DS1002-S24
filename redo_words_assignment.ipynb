```python
!pip install english_words
from english_words import get_english_words_set
words = get_english_words_set(['web2'], lower=True)
wordlist=list(words)
print("length of wordlist:",len(wordlist))
print("First 10 words:")
for word in wordlist[:10]:
    print(word)
# Q3-sorting these 10 words in alphabetical order
word_sort = ["netful", "earl", "fiendhead", "infectant", "fumy", "lavatera", "transorbital", "disentitle", "hypnesthesis", "subgrade"]
alpha_list = sorted(word_sort)
# Printing  the sorted list
print("words sorted in alphabetical order:")
for word in alpha_list:
    print(word)
#4)- finding number of words starting with the letter 'a'-0 indicates starting letter with index 0 - reference: https://stackoverflow.com/questions/39497934/find-how-many-words-start-with-certain-letter-in-a-list
count_start_with_a = sum(1 for word in words if word.startswith('a'))
print("Number of words starting with 'a':", count_start_with_a)
 #5-part 1- finding number of words that contain the letter 'q'
q_containing_words = sum(1 for word in words if 'q' in word)
print("Number of words containing 'q':", q_containing_words)
#5-part 2- finding the fraction of words that do not contain the letter 'e'
# first, finding number of words that do not contain 'e'
words_without_e = sum(1 for word in words if 'e' not in word)

# to find the fraction, divide by total number of words
fraction_without_e = words_without_e / len(words)

print("Fraction of words without 'e':", fraction_without_e)
#Q6
import numpy as np
#converting lists to numpy arrays
worldlist_np=np.array(wordlist)
alpha_list_np=np.array(alpha_list)
alphabet = [chr(i) for i in range(ord('a'),ord('z')+1)]
print(alphabet)
#this code generates a list called alphabet which contains all letters from 'a' to 'z' in alphabetical order

#7making a bar graph of the number of words starting with each letter
import matplotlib.pyplot as plt

# dictionary method learnt in class to store word counts beginning with each letter
word_counts = {letter: 0 for letter in alphabet}

# going through each word in alpha_list and update the counts
for word in wordlist:
    first_letter = word[0]
    word_counts[first_letter] += 1

# Extract letters and corresponding counts
letters = list(word_counts.keys()) # retrieves all the letters (keys) from the word_counts dictionary
counts = list(word_counts.values()) #retrives values which represents number of words starting with the corresponding letter

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(letters, counts, color='red')
plt.xlabel('Letters')
plt.ylabel('Number of Words')
plt.title('Number of Words Starting with Each Letter')
plt.show()

#8- finding the six longest words in the list
# Sort the wordlist by word length in descending order (longist to shortist word)
sorted_wordlist = sorted(wordlist, key=len, reverse=True)

# Select the top 6 longest words
six_longest_words = sorted_wordlist[:6]

# Print the  6 longest words
print("The 6 longest words in wordlist:")
for word in six_longest_words:
    print(word)

#Q9- making a histogram of the number of words with each length
import matplotlib.pyplot as plt
import numpy as np
# Create a dictionary to store the count of words for each length
word_length_counts = {}

# Iterate through each word in wordlist and update the counts
for word in wordlist:
    length = len(word)
    if length in word_length_counts:
        word_length_counts[length] += 1
    else:
        word_length_counts[length] = 1

# Extract lengths and corresponding counts
lengths = list(word_length_counts.keys())
counts = list(word_length_counts.values())

# Create the histogram
plt.figure(figsize=(10, 6))
plt.bar(lengths, counts, color='seagreen')
plt.xlabel('Word Length')
plt.ylabel('Number of Words')
plt.title('Histogram of Word Lengths in wordlist')
plt.show()

# Find the most common length of a word
most_common_length = max(word_length_counts, key=word_length_counts.get)
print("The most common length of a word in the wordlist is:", most_common_length,"letters")

#Q11 part 1- how many words contain the letters 'a' and 'v'--- executed in 15 seconds
# creating a variable for words containing both 'v' and 'a'
words_with_va = 0

# Iterate through each word in wordlist
for word in wordlist:
    # Check if both 'v' and 'a' are present in the word
    if 'v' in word and 'a' in word:
        words_with_va += 1

print("Number of words containing both 'v' and 'a':", words_with_va)
#number of times looped through wordlist to find words containing both 'v' and 'a'
# Get the length of wordlist
num_iterations = len(wordlist)

print("Number of iterations through wordlist:", num_iterations)
#Q11 part 2- how many words contain both the letters 'w' and 'l'--- took 15 seconds to run

words_with_wl = 0
#iterate through each word in wordlist
for word in wordlist:
  #check if both 'w' and 'l' are present in the word
  if 'w' in word and 'l'in word:
    words_with_wl+=1
print("Number of words containing both 'w' and 'l':", words_with_wl)
#number of times looped through wordlist to find words containing both 'w' and 'l'
#get the length of wordlist
num_iterations=len(wordlist)
print("number of iterations through wordlist:", num_iterations)


#bonus point general notes:

#bonus point - most frequebt letter pair
# Initialize a dictionary to store counts of pairs of letters
letter_pairs_count = {}

# Iterate through each word in wordlist
for word in wordlist:
    # Iterate through all possible pairs of letters in the word
    for i in range(len(word)):
        for j in range(i+1, len(word)):
            # Extract the pair of letters
            pair = tuple(sorted([word[i], word[j]]))
            # Increment the count for the pair in the dictionary
            if pair in letter_pairs_count:
                letter_pairs_count[pair] += 1
            else:
                letter_pairs_count[pair] = 1

# Find the pair of letters that appear in the most words
most_common_pair = max(letter_pairs_count, key=letter_pairs_count.get)
most_common_count = letter_pairs_count[most_common_pair]

print("Pair of letters that appear in the most words:", most_common_pair)
print("Number of words containing this pair of letters:", most_common_count)
#bonus point- least frequent word pair
# Initialize variables to store the pair with the fewest words and its count
fewest_words_pair = None
fewest_words_count = float('inf')

# Iterate through each pair of letters in the nested loops
for i in range(len(alphabet)):
    for j in range(i+1, len(alphabet)):
        pair = (alphabet[i], alphabet[j])
        # Check if the pair exists in the letter_pairs_count dictionary
        if pair in letter_pairs_count:
            # Update fewest_words_pair and fewest_words_count if the count is lower
            if letter_pairs_count[pair] < fewest_words_count:
                fewest_words_count = letter_pairs_count[pair]
                fewest_words_pair = pair

print("Pair of letters contained in the fewest words:", fewest_words_pair)
print("Number of words containing this pair of letters:", fewest_words_count)

#bonus point part 2- heat map
#using seaborn library https://seaborn.pydata.org
import seaborn as sns
import matplotlib.pyplot as plt

# Create a matrix to store the frequencies of pairs of letters
heatmap_data = []

# Iterate through each pair of letters in the nested loops
for i, letter1 in enumerate(alphabet):
    row = []
    for j, letter2 in enumerate(alphabet):
        pair = (letter1, letter2)
        # Get the frequency of the pair from letter_pairs_count
        frequency = letter_pairs_count.get(pair, 0)
        row.append(frequency)
    heatmap_data.append(row)

# Create a heatmap using Seaborn
plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='d', xticklabels=alphabet, yticklabels=alphabet)
#cmap='viridis': This parameter specifies the colormap to be used for coloring the cells in the heatmap.  'viridis' is a colormap that goes from yellow to green to blue- commonly used for visualizing numerical data.
plt.title('Heatmap of Pair Frequencies')
plt.xlabel('Second letter')
plt.ylabel('First letter')
plt.show()
```
